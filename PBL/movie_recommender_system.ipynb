```python
import numpy as np
import pandas as  pd
import re
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
movies.head()
```

# DATA CLEANING AND TRASFORMATION

```python
movies.head(1)
credits.head(1)['crew'].values
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.columns
```

```python
movies=movies[['movie_id','genres','keywords','title','overview','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast #abstract syntax tree
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
    #l=re.findall('(?:"name": ")(\w*)',obj)
    #return l
```

```python
#appropriating values of genres and keywords column
movies['genres']=movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['cast'][0]
```

```python
#to solve the cast values(selecting only 3 people in cast)
def convert3(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())#splits the overview column values and place them as dictionaries
movies.head()
```

```python
#we have to remove spaces between the names because if there are 2 person or genres with same first name or last name, then both genres or persons movie will be recommending

movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
import nltk #natural language toolkit
```

```python
from nltk.stem.porter import PorterStemmer#used for removing similar word by basic word
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df.head()
```

# VECCTORIZATION PROCESS
Process of converting given data into vectors so that closest one around this vector can be recommended

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
len(cv.get_feature_names())
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
#building a similarity matrix that is having values after comparing with each of the vectors including itself
```

```python
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    print(movies_list)
    for i in movies_list:
        print(new_df.iloc[i[0]]['title'])
    
    
    
```

```python
recommend('Harry Potter and the Half-Blood Prince')
```
